```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
test_data = pd.read_csv('testData.csv')
train_data = pd.read_csv('trainingData.csv')

# Convert the timestamp column to datetime
train_data['timestamp'] = pd.to_datetime(train_data['timestamp'])
test_data['timestamp'] = pd.to_datetime(test_data['timestamp'])

# Set the timestamp column as the index
train_data.set_index('timestamp', inplace=True)
test_data.set_index('timestamp', inplace=True)
```

```python
# Plot the time series data for pressure_1, pressure_2, and pressure_3
plt.figure(figsize=(14, 7))

# Train data
plt.plot(train_data.index, train_data['pressure_1'], label='Train Pressure 1')
plt.plot(train_data.index, train_data['pressure_2'], label='Train Pressure 2')
plt.plot(train_data.index, train_data['pressure_3'], label='Train Pressure 3')

# Test data
plt.plot(test_data.index, test_data['pressure_1'], label='Test Pressure 1', linestyle='--')
plt.plot(test_data.index, test_data['pressure_2'], label='Test Pressure 2', linestyle='--')
plt.plot(test_data.index, test_data['pressure_3'], label='Test Pressure 3', linestyle='--')

plt.xlabel('Timestamp')
plt.ylabel('Pressure')
plt.title('Time Series Visualization of Pressure 1, 2, and 3')
plt.legend()
plt.show()
```

```python
# Plot the time series data for temperature_1, temperature_2, temperature_3, and temperature_4
plt.figure(figsize=(14, 7))

# Train data
plt.plot(train_data.index, train_data['temperature_1'], label='Train Temperature 1')
plt.plot(train_data.index, train_data['temperature_2'], label='Train Temperature 2')
plt.plot(train_data.index, train_data['temperature_3'], label='Train Temperature 3')
plt.plot(train_data.index, train_data['temperature_4'], label='Train Temperature 4')

# Test data
plt.plot(test_data.index, test_data['temperature_1'], label='Test Temperature 1', linestyle='--')
plt.plot(test_data.index, test_data['temperature_2'], label='Test Temperature 2', linestyle='--')
plt.plot(test_data.index, test_data['temperature_3'], label='Test Temperature 3', linestyle='--')
plt.plot(test_data.index, test_data['temperature_4'], label='Test Temperature 4', linestyle='--')

plt.xlabel('Timestamp')
plt.ylabel('Temperature')
plt.title('Time Series Visualization of Temperature 1, 2, 3, and 4')
plt.legend()
plt.show()
```

```python
# Plot the time series data for vibration
plt.figure(figsize=(14, 7))

# Train data
plt.plot(train_data.index, train_data['vibration'], label='Train Vibration')

# Test data
plt.plot(test_data.index, test_data['vibration'], label='Test Vibration', linestyle='--')

plt.xlabel('Timestamp')
plt.ylabel('Vibration')
plt.title('Time Series Visualization of Vibration')
plt.legend()
plt.show()
```

```python
# Plot the time series data for condition
plt.figure(figsize=(14, 7))

# Train data
plt.plot(train_data.index, train_data['condition'], label='Train Condition')

plt.xlabel('Timestamp')
plt.ylabel('Condition')
plt.title('Time Series Visualization of Condition')
plt.legend()
plt.show()
```
